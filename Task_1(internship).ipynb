```python
df3=pd.read_csv('/content/netflix_titles.csv')
df3.info()
df3
```

```python
df3.head()
df3.info()
df3.isnull().sum()
```

```python
df3['director'].fillna('Unknown',inplace=True)
df3['cast'].fillna('Unknown',inplace=True)
df3['country'].fillna('Unknown',inplace=True)
df3
```

```python
mode_rating = df3['rating'].mode()
if not mode_rating.empty:
    fill_value = mode_rating.iloc[0]
else:
    fill_value = 'Unknown'
df3.loc[:,'rating']=df3['rating'].fillna(fill_value)
```

```python
print(df3['rating'].value_counts(dropna=False))
print(df3['rating'].mode())
```

```python
df3['duration']=df3['duration'].fillna('Unknown')
```

```python
df3['date_added']=pd.to_datetime(df3['date_added'],errors='coerce')
df3['date_added']=df3['date_added'].fillna(df3['date_added'].mode()[0])
```
